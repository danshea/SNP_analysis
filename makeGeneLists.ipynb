```python
import pandas as pd
```

```python
S11='S11-homozygous-variants.snpeff.unique.HIGH.tsv'
R09='R09-homozygous-variants.snpeff.unique.HIGH.tsv'
Shared='R09-S11-homozygous-variants.snpeff.shared.HIGH.tsv'
```

```python
S11df = pd.read_table(S11, header=None)
R09df = pd.read_table(R09, header=None)
Shareddf = pd.read_table(Shared, header=None)
```

```python
def writeGeneLists(df, filename):
    isINDEL = False
    with open(filename, 'w') as fh:
        for (index, row) in df.iterrows():
            chromosome = row[0]
            position   = row[1]
            reference  = row[2]
            alternative= row[3]
            info       = row[4]
            # Parse the info entry
            # First split it on ';'
            info_list = info.split(';')
            if info[0] == 'INDEL':
                isINDEL = True
            else:
                isINDEL = False
            for i in info_list:
                if i.startswith('ANN='):
                    ANN = i
                    break
            # Next parse the ANN
            ANN_list = ANN.split('|')
            effects = ANN_list[1]
            impact  = ANN_list[2]
            gene    = ANN_list[3]
            
            # Write out the details as tab-delimited format
            outline = '\t'.join([chromosome, str(position), reference, alternative, 
                                 'INDEL' if isINDEL else 'SNP', effects.replace('&',','), impact, gene])
            fh.write('{}\n'.format(outline))
            
```

```python
writeGeneLists(S11df, 'S11-homozygous-variants.snpeff.unique.HIGH.parsed.tsv')
```

```python
writeGeneLists(R09df, 'R09-homozygous-variants.snpeff.unique.HIGH.parsed.tsv')
```

```python
writeGeneLists(Shareddf, 'R09-S11-homozygous-variants.snpeff.shared.HIGH.parsed.tsv')
```
